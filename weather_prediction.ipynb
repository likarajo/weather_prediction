## Import Data

```python
import pandas as pd

data = pd.read_csv('data/data_ready.csv')

data.head(data.shape[0])
```

## Split dat into Training and Test sets

```python
from sklearn.model_selection import train_test_split

X = data.drop(['meanTemp'], axis=1)
y = data['meanTemp']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

## Create and Train Model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train, y_train)
```

## Evaluate the Model

```python
from sklearn.metrics import mean_absolute_error, median_absolute_error

prediction = regressor.predict(X_test)

score = regressor.score(X_test, y_test)*100
print('Explained Variance: %.2f' % score,'%')

mean_error = mean_absolute_error(y_test, prediction)
print('Mean Absolute Error: %.2f units' % mean_error)

median_error = median_absolute_error(y_test, prediction)
print('Median Absolute Error: %.2f units' % median_error)
```
