## Import Dataset

```python
import pandas as pd

data = pd.read_csv('data/data_preprocessed.csv')

data.head(data.shape[0])
```

## Data Analysis

### Find the correlation of the dependant variable with each independant variable

```python
data.corr()[['meanTemp']].sort_values('meanTemp')
```

## Keep relevant features

```python
predictors = [
    'humidity','windSpeed','ozone','temperatureHigh_-1','temperatureLow_-1',
    'humidity_-1','pressure_-1','dewPoint_-1','windGust_-1','ozone_-1',
    'humidity_-2','pressure_-2','dewPoint_-2','ozone_-2','temperatureHigh_-2',
    'precipProbability_-2','cloudCover_-2','uvIndex_-2'
]

final_data = data[['meanTemp'] + predictors]

final_data.to_csv('data/data_ready.csv', index=False)
```
