## Data Collection

```python
import datetime as dt
from datetime import datetime
from collections import namedtuple
import requests

BASE_URL = "https://api.darksky.net/forecast/{}/{},{}?exclude=[currently,minutely,hourly,alerts,flags]&units=auto"
API_KEY = "332ea12fe1ec403355a3f785f4d9b061"

lat = 22.5726
lon = 88.3639

features = ["date","summary","cloudCover","temperatureHigh","temperatureLow",
    "humidity","precipProbability","precipType","dewPoint","pressure",
    "windSpeed","windGust","visibility","uvIndex","ozone"]

DailySummary = namedtuple("DailySummary", features)

def get_weather_data(url, api_key, today):
    records = []
    request = BASE_URL.format(api_key,str(lat),str(lon))
    response = requests.get(request)
    if response.status_code == 200:
        for i in range(8):
            data = response.json()['daily']['data'][i]
            _date = today+dt.timedelta(days=i)
            records.append(
                DailySummary(
                    date = str(_date),
                    summary = data['summary'],
                    cloudCover = data['cloudCover'],
                    temperatureHigh = data['temperatureHigh'],
                    temperatureLow = data['temperatureLow'],
                    humidity = data['humidity'],
                    precipProbability = data['precipProbability'],
                    precipType = data['precipType'],
                    dewPoint = data['dewPoint'],
                    pressure = data['pressure'],
                    windSpeed = data['windSpeed'],
                    windGust = data['windGust'],
                    visibility = data['visibility'],
                    uvIndex = data['uvIndex'],
                    ozone = data['ozone']
                )
            )
    return records

today = dt.date.today()

records = get_weather_data(BASE_URL, API_KEY, today)
```

## Import Data

```python
import pandas as pd

df = pd.DataFrame(records, columns=features).set_index('date')

df.head(df.shape[0])
```

## Save Data

```python
now = dt.datetime.now()
current_time = now.strftime("%Y-%m-%d_%H:%M")
filename = 'data/'+current_time+'.csv'

df.to_csv(filename, index=False)
print('Data saved in', filename)
```
